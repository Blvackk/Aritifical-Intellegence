```python
import spacy
from spacy.lang.en.stop_words import STOP_WORDS
from string import punctuation
```

```python
stopwords = list(STOP_WORDS)
stopwords
```

```python
text
```

```python
doc = nlp(text)
```

```python
doc
```

# Docs to Tokens 

```python
# lets get the tokens from text
tokens = [token.text for token in doc]
print(tokens)
```

```python
doc
```

```python
# we have to calculate the frequency of each and every token, how many time token is repeated in text
word_frequencies = {}
for word in doc:
    if word.text.lower() not in stopwords and word.text.lower() not in punctuation:
        if word.text.lower() not in word_frequencies.keys():
            word_frequencies[word.text.lower()] = 1
        else:
            word_frequencies[word.text.lower()] += 1
```

```python
word_frequencies
```

```python
len(word_frequencies)
```

```python
max_frequency = max(word_frequencies.values())
max_frequency
```

```python
# to get normalized/weighted frequencies you should divide each frequency by the maximum frequency. This will give you a value between 0 and 1 for each token, indicating its relative importance in the text.

for word in word_frequencies.keys():
    word_frequencies[word] = word_frequencies[word]/max_frequency
```

```python
word_frequencies
```

# Docs to Sentences 

```python
sentences_tokens= [sent for sent in doc.sents]
sentences_tokens
```

```python
len(sentences_tokens)
```

```python
sentences_tokens
```

```python
#we are going to calculate the score of each sentence by adding the frequency of each word in the sentence. This will give us a score for each sentence, which we can use to rank the sentences and select the top sentences for our summary.

sentence_scores = {}

for sent in sentences_tokens:
    for word in sent:
        if word.text.lower() in word_frequencies.keys():
            if sent not in sentence_scores.keys():
                sentence_scores[sent] = word_frequencies[word.text.lower()]
            else:
                sentence_scores[sent] += word_frequencies[word.text.lower()]
```

```python
sentence_scores
```

```python
len(sentence_scores)
```

```python
from heapq import nlargest
```

```python
select_length = int(len(sentences_tokens)*0.4)
select_length
```

```python
# we have to select maximum 40% sentences out of all setences
summary = nlargest(select_length, sentence_scores, key=sentence_scores.get)
summary
```

```python
# if i need to combine these top 3 sentences then
final_summary = [word.text for word in summary]
final_summary
```
