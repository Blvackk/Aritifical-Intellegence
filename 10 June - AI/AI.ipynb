```python
import os
import nltk
# nltk.download()
```

```python
AI = '''Artificial Intelligence refers to the intelligence of machines. This is in
humans and animals. With Artificial Intelligence, machines perform functions such a
problem-solving. Most noteworthy, Artificial Intelligence is the simulation of huma
It is probably the fastest-growing development in the World of technology and innov
AI could solve major challenges and crisis situations.'''
```

```python
AI
```

```python
type(AI)
```

```python
from nltk.tokenize import word_tokenize
```

```python
ai_tokens=word_tokenize(AI)
```

```python
print(ai_tokens)
```

```python
len(ai_tokens)
```

```python
from nltk.tokenize import sent_tokenize
```

```python
ai_sent=sent_tokenize(AI)
```

```python
print(ai_sent)
```

```python
len(ai_sent)
```

```python
from nltk.tokenize import blankline_tokenize
ai_blank=blankline_tokenize(AI)
print(ai_blank)
```

```python
len(ai_blank)
```

```python
from nltk.tokenize import WhitespaceTokenizer
wt=WhitespaceTokenizer().tokenize(AI)
```

```python
print(wt)
```

```python
len(wt)
```
